```python
import pandas as pd
dados = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")

votos_acima_de_100 = dados[dados['vote_count'] >= 100]
votos_acima_de_100
```

```python
import numpy as np
import pandas as pd


contagem_votos_media = pd.cut(votos_acima_de_100['vote_average'], bins=np.arange(0, 11, 1), right=False, include_groups=False).value_counts()


agrupando_notas = {interval: int(interval.left) for interval in contagem_votos_media.index}
contagem_votos_media.index = contagem_votos_media.index.map(agrupando_notas)

contagem_votos_media_df = contagem_votos_media.reset_index()
contagem_votos_media_df.columns = ['nota', 'numero_de_filmes']
contagem_votos_media_df

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data= contagem_votos_media_df, x='nota', y='numero_de_filmes', order = contagem_votos_media_df['nota'])

plt.title('DistribuiÃ§Ã£o de Filmes por Notas')
plt.xlabel('Notas')
plt.ylabel('NÃºmero de Filmes')
```
